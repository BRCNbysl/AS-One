```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from transformers import pipeline

# Modeli yükle
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Pipeline ile metin sınıflandırma
classifier = pipeline("text-classification", model=model, tokenizer=tokenizer)

# Test verisi
texts = ["I love programming.", "This is a terrible idea."]
results = classifier(texts)

# Sonuçları yazdır
for text, result in zip(texts, results):
    print(f"Metin: {text}")
    print(f"Etiket: {result['label']}, Skor: {result['score']:.4f}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Veriyi yükle
data = pd.read_csv("help_desk_requests.csv")

# Metin ve etiket sütunlarını al
texts = data["Talepler"].tolist()
labels = data["Etiket"].tolist()

# Etiketleri sayısallaştır (örneğin, Teknik Servis = 0, IT Destek = 1)
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)

# Eğitim ve test verisini ayır
train_texts, val_texts, train_labels, val_labels = train_test_split(
    texts, encoded_labels, test_size=0.2, random_state=42
)
```

```python
from transformers import AutoTokenizer

# Tokenizer'ı yükle
model_name = "roberta-base"  # Hafif ve hızlı bir model
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Metinleri tokenize et
train_encodings = tokenizer(train_texts, truncation=True, padding=True, max_length=128)
val_encodings = tokenizer(val_texts, truncation=True, padding=True, max_length=128)

# Tensor veri seti oluştur
import torch

class Dataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item["labels"] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = Dataset(train_encodings, train_labels)
val_dataset = Dataset(val_encodings, val_labels)
```

```python
from transformers import AutoModelForSequenceClassification

# Modeli yükle (Etiket sayısını belirt)
num_labels = len(label_encoder.classes_)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=30)
```

```python
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",  # Güncellenmiş parametre adı
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=5,
    weight_decay=0.01,
    save_steps=10_000,
    save_total_limit=2,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
)

# Modeli eğit
trainer.train()

```

```python
model.save_pretrained("./helpdesk_model")
tokenizer.save_pretrained("./helpdesk_model")
```

```python
# Test verisi
test_text = "Yazıcımdan çıktı alamıyorum."
test_encoding = tokenizer(test_text, truncation=True, padding=True, return_tensors="pt")

# Model ile tahmin yap
model.eval()
with torch.no_grad():
    outputs = model(**test_encoding)
    logits = outputs.logits
    predicted_class = torch.argmax(logits, dim=1).item()

# Sonuçları çöz
predicted_label = label_encoder.inverse_transform([predicted_class])
print(f"Tahmin edilen sınıf: {predicted_label[0]}")
```
